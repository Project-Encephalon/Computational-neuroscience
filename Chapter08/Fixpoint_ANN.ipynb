```python
#fixpoint ANN model of an auto-associative neural network

#importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt


rand=np.random.random((500,10))           #random number generation for patterns
rand_x=np.random.random((500,1))          #random number generation for patterns 
pat=2*np.floor(2*rand)-1                  #Random 10 pattern of network with 500 nodes
w=np.matmul(pat, np.transpose(pat))       #Hebbian learning
l=rand_x-0.5                              #Initialization of network
m=np.zeros((500, 9))
s=np.concatenate((l,m), axis=1)           #initial state of the network in the range of −0.5 ≤ s < 0.5.


for t in range (1,10):
  s[:,t]=np.sign(np.matmul(w,s[:,t-1]))   #updates the network for nine time steps with a loop


s_tpose=np.transpose(s)                   #Transposing S
pat=pat/500                               #Normalisation of values
x=np.matmul(s_tpose,pat)                  

```

```python
#Plotting and labelling of results

plt.plot(x)                               
plt.title("Fixpoint ANN Model")
plt.ylabel("Overlap")
plt.xlabel("Iterations")
plt.show()
```
